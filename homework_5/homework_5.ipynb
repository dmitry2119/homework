### Функция одной переменной

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 101)
f1 = lambda x: 3 * x ** 2
y = f1(x)
```

```python
plt.plot(x, y)

plt.grid(True)
```

```python
df1 = lambda x: 6 * x
```

```python
x0 = 5
lam = 0.1
```

```python
res = []
x1 = x0 - lam * df1(x0)
while np.linalg.norm(x0 - x1) > 0.00000001:
    res.append([x0, f1(x0)])
    x0 = x1
    x1 = x0 - lam * df1(x0)
```

```python
res = np.array(res)
res
```

```python
plt.plot(x, y)
plt.plot(res[:, 0], res[:, 1], 'o-', c='r')

plt.grid(True)
```

### Функция двух переменных

```python
f2 = lambda x, y: 3 * x ** 2 + 5 * y ** 2
```

```python
x = np.linspace(-5, 5, 101)
y = np.linspace(-5, 5, 101)
```

```python
x, y = np.meshgrid(x, y)
z = f2(x, y)
z
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(x, y, z)
```

```python
def df2(x, y):
    return np.array([6 * x, 10 * y])
```

```python
x0 = np.array([5, 10])
lam = 0.1
```

```python
res = []
x1 = x0 - lam * df2(*x0)
while np.linalg.norm(x0 - x1) > 0.00000001:
    res.append([*x0, f2(*x0)])
    x0 = x1
    x1 = x0 - lam * df2(*x0)
```

```python
res = np.array(res)
res
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(x, y, z)
ax.plot(res[:, 0], res[:, 1], res[:, 2], 'o-', c='r')

plt.grid(True)
```
